```python
from skimage import data, draw
from skimage import transform, util
import numpy as np
from skimage import filters, color
from matplotlib import pyplot as plt


hl_color = np.array([0, 1, 0])

#Loading Data

img = data.chelsea()
img = util.img_as_float(img)

#Using Sobel Filter

eimg = filters.sobel(color.rgb2gray(img))

plt.title('Original Image')
plt.imshow(img)

#Resizing to a new aspect ratio distorts image contents.

resized = transform.resize(img, (img.shape[0], img.shape[1] - 200),
                           mode='reflect')
plt.figure()
plt.title('Resized Image')
plt.imshow(resized)

#Seam carving attempts to resize without distortion, by removing regions of an image which are less important

out = transform.seam_carve(img, eimg, 'vertical', 200)
plt.figure()
plt.title('Resized using Seam Carving')
plt.imshow(out)
```
